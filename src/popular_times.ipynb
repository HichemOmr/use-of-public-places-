```python
### Testing for exporting excel file
import populartimes
import pandas as pd

# Function to get popular times data
def get_popular_times(api_key, place_id):
    try:
        data = populartimes.get_id(api_key, place_id)
        return data
    except Exception as e:
        print(f"Error retrieving data: {e}")
        return None

# Function to process data and export to Excel
def process_and_export(data, file_path):
    if not data:
        print("No data to process.")
        return

    # Extract and format the data
    days = data.get('populartimes', [])
    if not days:
        print("No popular times data available.")
        return

    df = pd.DataFrame(days)
    hourly_df = df['data'].apply(pd.Series)
    hourly_df.columns = [f'Hour_{i}' for i in range(24)]
    final_df = pd.concat([df[['name']], hourly_df], axis=1)
    final_df = final_df.rename(columns={'name': 'Day'})

    # Export to Excel
    final_df.to_excel(file_path, index=False)
    print(f'Data exported to {file_path}')

# Main script
if __name__ == "__main__":
    API_KEY = "API"  # Replace with your Google Maps API key
    PLACE_ID = "ChIJSRAEZNVOlUcRV_BppDN4Kmg"  # Replace with your place ID
    EXCEL_PATH = 'popular_times_recent.xlsx' # Replace with the name you want

    # Get data
    popular_times_data = get_popular_times(API_KEY, PLACE_ID)

    # Process and export data
    process_and_export(popular_times_data, EXCEL_PATH)
```

```python
import populartimes
import pandas as pd
import os
import time

# Replace with your API key
api_key = "API Key"

# Load the Excel file with place IDs
input_file = "../data/SusDens-Database (Final).xlsx"  # Replace with your actual file name if you want to update the data (i.e: have more places)
place_ids_df = pd.read_excel(input_file)

# Ensure the main output directory exists
output_directory = "../out/popular_times/" ### change the name if you want
os.makedirs(output_directory, exist_ok=True)

# Ensure the places directory exists
places_directory = os.path.join(output_directory, 'places') ### change the name if you want
os.makedirs(places_directory, exist_ok=True)

# Initialize lists to keep track of processed data
places_with_data = []
places_without_data = []
errors = []

# Define hour columns for output
hour_columns = [f"Hour_{i}" for i in range(24)]

# Iterate through each place ID in the DataFrame
for index, row in place_ids_df.iterrows():
    place_id = row['place_id']
    try:
        # Retrieve popular times data
        data = populartimes.get_id(api_key, place_id)
        
        # Debug: Print the retrieved data
        print(f"Data for Place ID {place_id}: {data}")
        
        # Extract the required information
        place_name = data.get('name', f'place_{index}')  # Use a fallback name if name is missing
        popular_times = data.get('populartimes', [])
        
        # Check if populartimes data is empty
        if not popular_times:
            print(f"No popular times data found for {place_name}")
            places_without_data.append({'place_id': place_id, 'place_name': place_name})
            continue  # Skip to the next place if no data
        
        # Prepare the popular times data for DataFrame
        formatted_data = []
        for day in popular_times:
            day_name = day['name']
            times_data = day['data']
            
            # Ensure we have exactly 24 hours of data, filling with zeros if necessary
            if len(times_data) < 24:
                times_data.extend([0] * (24 - len(times_data)))
            
            # Create a row with day_name and times_data
            formatted_data.append([day_name] + times_data)
        
        # Convert formatted data to DataFrame
        popular_times_df = pd.DataFrame(formatted_data, columns=['Day'] + hour_columns)
        
        # Replace invalid characters in the place name to avoid file system errors
        safe_place_name = "".join([c if c.isalnum() or c in (' ', '_') else '_' for c in place_name])
        
        # Create a directory for the place
        place_directory = os.path.join(places_directory, safe_place_name)
        os.makedirs(place_directory, exist_ok=True)
        
        # Define the output file path
        output_file = os.path.join(place_directory, f"{safe_place_name}.xlsx")
        
        # Save the popular times data to an Excel file inside the folder
        popular_times_df.to_excel(output_file, index=False)
        
        print(f"Saved data for {place_name} in {output_file}")
        
        # Record place with data
        places_with_data.append({'place_id': place_id, 'place_name': place_name})
        
    except Exception as e:
        # Handle any exceptions (e.g., place not found, quota exceeded)
        print(f"Error retrieving data for place ID {place_id}: {e}")
        errors.append({'place_id': place_id, 'error': str(e)})

    # Optional: sleep to respect rate limits
    time.sleep(1)  # Adjust sleep time as necessary

# Save results to Excel files
pd.DataFrame(places_with_data).to_excel(os.path.join(output_directory, "places_with_data.xlsx"), index=False)
pd.DataFrame(places_without_data).to_excel(os.path.join(output_directory, "places_without_data.xlsx"), index=False)
pd.DataFrame(errors).to_excel(os.path.join(output_directory, "errors.xlsx"), index=False)

print("All data processed. Results saved.")
```
