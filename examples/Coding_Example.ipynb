```python
def check_table_schema(table_name):
    print(f"\nSchema of {table_name} table:")
    cursor.execute(f"PRAGMA table_info({table_name});")
    schema = cursor.fetchall()
    for column in schema:
        print(column)

def check_foreign_keys(table_name):
    print(f"\nForeign keys in {table_name} table:")
    cursor.execute(f"PRAGMA foreign_key_list({table_name});")
    foreign_keys = cursor.fetchall()
    for fk in foreign_keys:
        print(fk)

try:
    # Check the schema of the photos table
    check_table_schema("photos")
    # Check foreign keys in the photos table
    check_foreign_keys("photos")

    # Check the schema of the reviews table
    check_table_schema("reviews")
    # Check foreign keys in the reviews table
    check_foreign_keys("reviews")

    # Check the schema of the PopularTimes table
    check_table_schema("PopularTimes")
    # Check foreign keys in the PopularTimes table
    check_foreign_keys("PopularTimes")

except sqlite3.Error as e:
    print(f"An error occurred: {e}")
```
